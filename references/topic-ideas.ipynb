Die Daten kommt aus transfermarkt.de. Der Datensatz wurde von einem Kaggle User über ein nicht offiziellen transfermarkt [scraper](https://github.com/dcaribou/transfermarkt-scraper) gesammelt. ### Data characterisitcs

Die Daten aus Kaggle beinhalten alle wichtigen Informationen aus der Fussball Welt. Der Datensatz besteht aus mehreren CSV-Dateien mit Informationen zu Wettbewerben, Spielen, Vereinen, Spielern und Auftritten. Jede Datei enthält die Attribute der Entität und die IDs, die verwendet werden können, um sie miteinander zu joinen. 
Alle Datensätze gejoined entstehen hier mehr als 10 Spalten und 100 Reihen.

Allgemein kann man in den Daten Infos aus jedes einzelnes Spiel (Tore, gelbe/rote Karten, Schiedsrichter), Aufstellungen zu jedem verfügbaren Spiel (Startelf, Auswechsel), Wettbewerbe (welche es gibt), Bewertung eines Spieles auf Spieler Ebene (Tore, gelbe/rote Karten, Spielzeit), Marktwert der Spieler, genauere Infos zu einem Spielverlauf (wenn ein Spieler eine Karte bekam, dann wird beschrieben wieso und wann; genauso bei einem Tor eine genaurer Beschreibung vorliegt), alle Infos zu den einezelnen transfers (Gebühren, derzeitiger Marktwert, woher zu wohin), ausgiebige Infos zu jeden einzelnen Spieler (Größe, Alter, Position, Geburtsort, Füßigkeit, uvm) und allgemeine Infos auf Verein Ebene (erzielte Tore, erhaltene Gegentore, Ausländerrate, Mannschaftsgröße, Stadiumsgröße, Netto Transfer Bilanz).
